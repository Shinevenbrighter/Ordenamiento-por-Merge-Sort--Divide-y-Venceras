##Introducción
El ordenamiento es una de las operaciones fundamentales en ciencias de la computación, muchos algoritmos dependen de que los datos esten organizados para funcionar de manera eficiente.

El algoritmo merge sort es uno de los ejemplos más clásicos de divide y vencerás.

Su funcionamiento se basa en:

*  Dividir: separar la lista en dos mitades
*  Conquistar: ordenar cada mitad de forma recursiva
*  Combinar: fusionar las dos mitades ya ordenadas en una sola lista también ordenada

```python
%%writefile merge_sort.py
# Importar librería
import numpy as np

# Función para ordenar una lista usando for anidados
def ordenar_sublista(sublista):
    for i in range(len(sublista)):
        for j in range(i + 1, len(sublista)):
            if sublista[i] < sublista[j]:   # Comparar para ordenar de mayor a menor
                sublista[i], sublista[j] = sublista[j], sublista[i]
    return sublista

# Función merge para fusionar dos listas ya ordenadas
def merge_desc(left, right):
    resultado = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] >= right[j]:
            resultado.append(left[i])
            i += 1
        else:
            resultado.append(right[j])
            j += 1

    # Agregar lo que sobra
    resultado.extend(left[i:])
    resultado.extend(right[j:])
    return resultado

# Función para verificar si una lista está ordenada (ascendente, descendente o desordenada)
def verificar_orden(lista):
    if all(lista[i] <= lista[i+1] for i in range(len(lista)-1)):
        return "Ascendente"
    elif all(lista[i] >= lista[i+1] for i in range(len(lista)-1)):
        return "Descendente"
    else:
        return "Desordenada"

# Versión de Merge Sort híbrido optimizado
def merge_sort_hibrido(lista):
    if len(lista) <= 1:
        return lista

    # Paso 1: dividir
    mid = len(lista) // 2
    left = lista[:mid]
    right = lista[mid:]

    # Checar si las sublistas ya están ordenadas
    orden_left = verificar_orden(left)
    orden_right = verificar_orden(right)

    # Paso 2: evitar ordenar si ya están listas
    if orden_left == "Descendente":
        pass  # ya está en orden
    elif orden_left == "Ascendente":
        left.reverse()  # solo invertir
    else:
        left = ordenar_sublista(left)

    if orden_right == "Descendente":
        pass
    elif orden_right == "Ascendente":
        right.reverse()
    else:
        right = ordenar_sublista(right)

    # Paso 3: unir las dos mitades ordenadas
    return merge_desc(left, right)


def main():
    # Repetir varias pruebas
    for prueba in range(4):
        print(f"\nPrueba #{prueba + 1}")
        N = int(input("Ingrese el número de elementos: "))

        # Generar lista aleatoria entre 0 y 100
        x = np.random.randint(0, 100, N).tolist()
        print("Lista original:")
        print(x)

        orden_lista = verificar_orden(x)
        print(f"Orden de la lista completa: {orden_lista}")

        # Si ya está descendente → no hacer nada
        if orden_lista == "Descendente":
            lista_ordenada = x
        # Si está ascendente → invertir
        elif orden_lista == "Ascendente":
            lista_ordenada = list(reversed(x))
        else:
            lista_ordenada = merge_sort_hibrido(x)

        print("\nLista ordenada (mayor a menor):")
        print(lista_ordenada)


if __name__ == "__main__":
    main()
```

## Conclusión

**Merge Sort**  
Es muy efectivo frente a algoritmos cuadráticos cuando se trabaja con listas medianas o grandes, a diferencia de otros algoritmos simples.  

Además, es un algoritmo **estable** y puede adaptarse fácilmente a datos almacenados en archivos, etc.  

Limitaciones: requiere un **espacio adicional de memoria** para realizar las fusiones y puede ser **más lento** debido al costo de las llamadas recursivas.

- **Complejidad temporal:** O(n²)
- **Complejidad espacial:** O(n)  

---

**Fuerza bruta factorial**  
Este método genera y revisa **todas las permutaciones posibles** de la lista hasta encontrar la ordenada.  

- **Complejidad temporal:** O(n · n!)  
- **Complejidad espacial:** O(n) si se generan sobre la marcha, o O(n!) si se almacenan todas las permutaciones.  

Es el enfoque menos eficiente, ya que crece **explosivamente con el tamaño de la entrada**. Solo se utiliza como ejemplo académico o para listas muy pequeñas.
